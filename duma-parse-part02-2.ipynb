```python
import os
import pymorphy2
import pandas as pd
from string import punctuation
```

```python
morph = pymorphy2.MorphAnalyzer()
```

```python
def get_norm_text(x):
    """
    Gets raw text, returns text in normalised form
    """
    punct_to_exclude = punctuation + "—«»–"
    for p in punct_to_exclude:
        x = x.replace(p, " ")
    words = x.split()
    norm_words = [morph.parse(word)[0].normal_form for word in words]
    text = " ".join(norm_words)
    return text
```

```python
def filter_by_keywords(df, L):
    """
    df - data frame with texts
    L - list of key words
    Returns data frame with rows that 
    contain key words from L
    """
    filtered_by_keys = []
    for key in L:
        chosen = df[df["words_norm"].str.contains(key)]
        #if sum(df["words_norm"].str.contains(key)) != 0:
            #print(key)
        filtered_by_keys.append(chosen)
    final = pd.concat(filtered_by_keys)
    return final
```
