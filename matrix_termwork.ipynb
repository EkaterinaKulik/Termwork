```python
import pandas as pd
from itertools import combinations
from collections import defaultdict
import numpy as np
```

# Матрица соавторства

```python
co_sponsorship = pd.read_csv('7 BILLS SECTIONS.csv', delimiter=';', encoding='utf-8', on_bad_lines='skip', engine='python')
co_sponsorship.head()
```

```python
co_sponsorship.drop(columns = ["Session", 'Date', 'Class_Num', 'SectionLabel', 'logical'], inplace = True)
```

```python
filtered_df = co_sponsorship[(co_sponsorship['Class_Label'] == "Образование") & (co_sponsorship['Actor/Type'] == "депутат")]
filtered_df.head(5)
```

```python
co_sponsorship_counts = defaultdict(int)


for bill_id, group in filtered_df.groupby('Bill_ID'):
    ids = group['ID_dep'].tolist()
    for id1, id2 in combinations(ids, 2):
        co_sponsorship_counts[frozenset([id1, id2])] += 1


unique_ids = sorted(filtered_df['ID_dep'].unique())


co_sponsorship_matrix = pd.DataFrame(0, index=unique_ids, columns=unique_ids)


for pair, count in co_sponsorship_counts.items():
    id1, id2 = list(pair)
    co_sponsorship_matrix.at[id1, id2] = count
    co_sponsorship_matrix.at[id2, id1] = count


co_sponsorship_matrix
```

# Матрица регионов

```python
relevant_data = data[['FullName', 'RegionOfBirth']].dropna(subset=['RegionOfBirth']).copy()
relevant_data['FullName'] = relevant_data['FullName'].astype(str)


co_occurrence_counts = defaultdict(int)


for region, group in relevant_data.groupby('RegionOfBirth'):
    dep_names = group['FullName'].tolist()
    if len(dep_names) > 1:
        for dep1, dep2 in combinations(dep_names, 2):
            co_occurrence_counts[frozenset([dep1, dep2])] += 1


unique_names = sorted(relevant_data['FullName'].unique())


co_occurrence_matrix = pd.DataFrame(0, index=unique_names, columns=unique_names)


for pair, count in co_occurrence_counts.items():
    if len(pair) == 2: 
        dep1, dep2 = list(pair)
        co_occurrence_matrix.at[dep1, dep2] = count
        co_occurrence_matrix.at[dep2, dep1] = count


co_occurrence_matrix
```

# Матрица партийности (общая)

```python
num_deputies = data.shape[0]
intersection_matrix = pd.DataFrame(0, index=data['FullName'], columns=data['FullName'])


for i in range(num_deputies):
    for j in range(i, num_deputies):
        
        parties_i = data.loc[i, ['Party_1', 'Party_2', 'Party_3']].dropna().unique()
        parties_j = data.loc[j, ['Party_1', 'Party_2', 'Party_3']].dropna().unique()
        
       
        if set(parties_i) & set(parties_j):
            intersection_matrix.iloc[i, j] = 1
            intersection_matrix.iloc[j, i] = 1  

intersection_matrix
```

```python
intersection_matrix.to_csv('co_partymembership_matrix.csv', encoding='utf-8')
```

# Матрица партийности (для каждой отдельной партии)

```python
all_parties = pd.unique(data[['Party_1', 'Party_2', 'Party_3']].values.ravel('K'))
all_parties = [party for party in all_parties if pd.notna(party)]  


party_matrices = {party: pd.DataFrame(0, index=data['Id'], columns=data['Id']) for party in all_parties}


for i in range(data.shape[0]):
    for j in range(i, data.shape[0]):
        
        parties_i = set(data.loc[i, ['Party_1', 'Party_2', 'Party_3']].dropna().unique())
        parties_j = set(data.loc[j, ['Party_1', 'Party_2', 'Party_3']].dropna().unique())
        
        
        for party in all_parties:
            if party in parties_i and party in parties_j:
                party_matrices[party].iloc[i, j] = 1
                party_matrices[party].iloc[j, i] = 1  


for party, matrix in party_matrices.items():
    filename = f"intersection_matrix_{party}.csv"
    matrix.to_csv(filename, encoding='utf-8-sig')
```
